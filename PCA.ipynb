### Scaling Data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_pixels=scaler.fit_transform(pixels)
scaled_pixels
```

### PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_results=pca.fit_transform(scaled_pixels)
pca_results
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
fig, ax=plt.subplots(figsize=(8,6), dpi=100)
ax=sns.scatterplot(x=pca_results[:,0], y=pca_results[:,1], 
                hue=df["number_label"].values,
           palette="Set1")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
pca1 = PCA(n_components=3)
pca1_results=pca1.fit_transform(scaled_pixels)
pca1_results
```

```python
from mpl_toolkits import mplot3d
ax=plt.axes(projection="3d")
ax.scatter3D(pca1_results[:,0],
            pca1_results[:,1],
            pca1_results[:,2],
            c=df["number_label"])
```
